First, we will import the necessary modules for easy data processing.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize'] = [15, 10]
pd.options.display.max_rows = 10
```

Next we will read in our data using `pandas` and store the data in the variable `df` (for DataFrame). You can think of a DataFrame like a table or 2D list!

```python
df = pd.read_csv('stocks.csv', parse_dates=['Date'])
```

```python
df
```

## Question 1: Which company has the highest stock prices?
This question can be answered in several different ways, but the easiest way is probably to plot the average `High` stock price for each `Company` using a **bar plot**.

```python
# insert your plot code here
sns.barplot(x=df['Company'], y=df['High'])
plt.show()
```

Looks like Amazon is the most expensive stock to buy!
***
## Question 2: Is there any relationship between the Open price and the Close price?
In order to quick inspect relationships between two pieces of information, we can use a **scatter plot**. If we see an upward trend in our scatter points, we can say that there is a relationship between the two variables.

```python
# insert your plot code here
sns.regplot(x=df['Open'], y=df['Close'])
plt.show()
```

It looks like there is a strong relationship between `Open` and `Close` prices! As the open price increases (to the right on the x-axis) the close price also increase (going up the y-axis).
***
## Question 3: Is there a trend in High price for Amazon?
One of the best ways to spot a trend in data is to use a line plot! This is especially effective when we are looking at the trend over time. First we have to filter just the Amazon (AMZN) stock prices and then we can make a line plot with the `Date` on the x-axis and the `High` prices on the y-axis.

```python
# insert your filter code here
amazon = df.loc[df.Company == 'Amazon (AMZN)']
# insert your plot code here
sns.lineplot(x=amazon['Date'], y=amazon['High'])
plt.show()
```
