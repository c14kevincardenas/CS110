# Lesson 24 - Data Visualization
In this lesson we will walk through importing data and vizualizing it to easily and quickly answer some questions about our data. Data Visualization is an amazing technique to explore your data, answer questions about your data and summarize your analysis!
***
First, we will import the necessary modules for easy data processing.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize'] = [15, 10]
pd.options.display.max_rows = 10
```

Next we will read in our data using the `pandas` module that we call `pd` and store the data in the variable `df` (for DataFrame). You can think of a DataFrame like a table or 2D list!

```python
df = pd.read_csv('tech_giants.csv', parse_dates=['Date'])
```

```python
df
```

## Question 1: Which company has the most Revenue ($B)?
This question can be answered in several different ways, but the easiest way is probably to plot the average `Revenue ($B)` stock price for each `Company` using a **bar plot**.

```python
sns.barplot(x=df['Company'], y=df['Revenue ($B)'], ci=None)
plt.show()
```

Looks like Apple has been the most profitable company since 2006!
***
## Question 2: Is there any relationship between a company's Revenue ($B) and the Number of Employees?
In order to quick inspect relationships between two pieces of information, we can use a **scatter plot**. If we see an upward trend in our scatter points, we can say that there is a relationship between the two variables.

```python
sns.regplot(x=df['Number of Employees'], y=df['Revenue ($B)'], ci=None)
plt.show()
```

It looks like there is a strong relationship between `Number of Employees` and `Revenue ($B)`! However, there is one company that is growing very quickly in their `Number of Employees`, but not getting as much `Revenue ($B)`. Generally speaking for the five companies, as the number of employees increases (to the right on the x-axis) the revenue in billions of dollars also increases (going up the y-axis).
***
## Question 3: Is there a trend in Revenue ($B) for Amazon?
One of the best ways to spot a trend in data is to use a line plot! This is especially effective when we are looking at the trend over time. First we have to filter just `Amazon` and then we can make a line plot with the `Date` on the x-axis and the `Revenue ($B)` prices on the y-axis.

```python
# insert your filter code here
amazon = df.loc[df.Company == 'Amazon']
# insert your plot code here
sns.lineplot(x=amazon['Date'], y=amazon['Revenue ($B)'])
plt.show()
```

### BONUS! 
Interested in taking a look at the trend for all five companies? Run the code below!

```python
sns.lineplot(x=df['Date'], y=df['Revenue ($B)'], hue=df['Company'])
plt.show()
```

We can use the `hue` parameter to represent a third variable, `Company`. Our first two variables are on the x and y-axis, `Date` and `Revenue ($B)` respectively.
***
Interested in finding out which company was growing really fast in our scatter plot, but not seeing as much profits? Run the code below to find out!

```python
sns.scatterplot(x=df['Number of Employees'], y=df['Revenue ($B)'], hue=df['Company'])
plt.show()
```

Looks like is was Amazon!! Notice that we are using the `hue` parameter again in our plot functions to separate all of the data into smaller categories. In our case, we are separating our data based on the `Company`.
